```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
path = "/content/drive/MyDrive/data/Auto_MPG_data.csv"
data = pd.read_csv(path)
data
```

```python
data = np.array(data)
data
```

```python
# tron du lieu
index = np.arange(data.shape[0])
np.random.shuffle(index)
data = data[index]
```

```python
# tach input va output
X = data[:, 1:]
y = data[:,0]
```

```python
# chia thanh tap train, val , test
N = data.shape[0]
Train_size = 0.7
Val_size = 0.2

Train_sample = int(Train_size * N)
Val_sample = Train_sample + int (Val_size * N)

X_train = X[:Train_sample]
y_train = y[:Train_sample]

X_val = X[Train_sample: Val_sample]
y_val = y[Train_sample: Val_sample]

print(X_train)
```

```python
# Nomalize du lieu
m = X_train.mean(axis = 0)
s = X_train.std(axis = 0)
X_train = (X_train - m)/s
X_val = (X_val - m)/s
```

```python
MLP_Regression_model = tf.keras.Sequential([
    tf.keras.layers.Dense(units =64, input_shape =[9], activation ="relu"), # tang thu 1
    tf.keras.layers.Dense(units =64, activation ="relu" ), # tang thu 2
    tf.keras.layers.Dense(units =1)
])
opt = tf.keras.optimizers.SGD(learning_rate = 0.0003)
MLP_Regression_model.compile(optimizer = opt, loss = "mse",
                             metrics = [tfa.metrics.RSquare(),tf.keras.metrics.MeanAbsoluteError()])

MLP_Regression_history = MLP_Regression_model.fit(X_train, y_train, validation_data=(X_val, y_val), batch_size =16,epochs =200)
```

```python
import matplotlib.pyplot as plt

plt.plot(MLP_Regression_history.history["loss"], label="loss")
plt.plot(MLP_Regression_history.history["val_loss"], label="val_loss")
plt.ylim([0,40])
plt.xlabel("Epoch")
plt.ylabel("loss")
plt.legend() # chu thich hinh ve
plt.grid(True) # hien thi bieu do dang luoi
```
